```python
df.columns = ['text','split','label']
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Define the proportion for each set (e.g., 60% train, 20% test, 20% dev)
train_size = 0.6
test_size = 0.2
dev_size = 0.2

# Split the data
train, temp = train_test_split(df, test_size=1 - train_size, random_state=42)
test, dev = train_test_split(temp, test_size=test_size / (test_size + dev_size), random_state=42)

# Print the sizes of the resulting sets
print("Train set size:", len(train))
print("Test set size:", len(test))
print("Dev set size:", len(dev))
```

```python
train['split'] = 'train'
test['split'] = 'test'
dev['split'] = 'dev'
```

```python
concatenated_df = pd.concat([train, test, dev], axis=0)


concatenated_df.reset_index(drop=True, inplace=True)
```

```python
concatenated_df.to_csv('dataframe.tsv', sep='\t', index=False, encoding='utf-8')
```
